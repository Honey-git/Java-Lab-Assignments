```python
# Step 1: import required libraries

%pip install numpy pandas matplotlib seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Required libraries imported successfully.")
```

```python
# Step 2: Load dataset
%pip install scikit-learn
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Step 3: understand data types
print("\nDataset information: ")
print(df.info())
```

```python
# Step 4: Exploratory Data Analysis (EDA)
# 1. statistical Summary
df.describe()
```

```python
# 2. Scatter Plot (Two features)
plt.figure()
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# 3. Pair plot (all features)
sns.pairplot(df)
plt.show()
```

```python
# Step 5: Distance Measures Implementation

# select two data points
point1 = df.iloc[0].values
point2 = df.iloc[1].values
print("\nPoint 1:", point1)
print("\nPoint 2:", point2)

# a. Euclidean Distance
euclidean_distance = np.sqrt(np.sum((point1 - point2) ** 2))
print("\nEuclidean Distance: ", euclidean_distance)

# b. Manhattan Distance
manhattan_distance = np.sum(np.abs(point1 - point2))
print("\nManhattan Distance: ", manhattan_distance)

# c. Cosine Similarity
cosine_similarity = np.dot(point1, point2) / (np.linalg.norm(point1) * np.linalg.norm(point2))
print("\nCosine Similarity: ", cosine_similarity)
```

```python
# Step 6: Effect of Feature scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

print("\nOriginal Data (first row): ")
print(df.iloc[0].values)

print("\nScaled data (first row): ")
print(scaled_data[0])
```

```python
# distance measures on scaled data

point1 = scaled_data[0]
point2 = scaled_data[1]
print("\nPoint 1:", point1)
print("\nPoint 2:", point2)

# a. Euclidean Distance
euclidean_distance = np.sqrt(np.sum((point1 - point2) ** 2))
print("\nEuclidean Distance: ", euclidean_distance)

# b. Manhattan Distance
manhattan_distance = np.sum(np.abs(point1 - point2))
print("\nManhattan Distance: ", manhattan_distance)

# c. Cosine Similarity
cosine_similarity = np.dot(point1, point2) / (np.linalg.norm(point1) * np.linalg.norm(point2))
print("\nCosine Similarity: ", cosine_similarity)
```
