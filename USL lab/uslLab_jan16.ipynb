```python
%pip install numpy pandas sklearn scipy

import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from scipy.spatial import distance
```

```python
# Load sample dataset
iris = load_iris()
X = iris.data[:5] # Taking first 5 samples for demonstration
print("Sample data:\n", X)

# Select two sample points
point1 = X[0]
point2 = X[1]

# 1. Euclidean Distance
euclidean_dist = distance.euclidean(point1, point2)
print("Euclidean Distance: ", euclidean_dist)

# 2. Manhattan Distance
manhattan_dist = distance.cityblock(point1, point2)
print("Manhattan Distance: ", manhattan_dist)

# 3. Cosine Similarity
cosine_sim = 1 - distance.cosine(point1, point2)
print("Cosine Similarity: ", cosine_sim)

# 4. Minkowski Distance (p=3 for demostration)
minkowski_dist = distance.minkowski(point1, point2, 3)
print("Minkowski Distance (p=3): ", minkowski_dist)

# 5. Hamming Distance (binary example)
bin_1 = np.array([1, 0, 1, 1])
bin_2 = np.array([1, 1, 0, 1])
hamming_dist = distance.hamming(bin_1, bin_2)
print("Hamming Distance: ", hamming_dist)

```

```python
# Chessboard Distance Demonstration
pointA = (2, 3)
pointB = (5, 7)
chessboard_dist = distance.chebyshev(pointA, pointB)
print("Chessboard Distance: ", chessboard_dist)
```

```python
# Jaccarad Distance Demonstration

# Example binary vectors
set1 = [1, 0, 1, 1, 0]
set2 = [1, 0, 0, 1, 1]

jaccard_dist = distance.jaccard(set1, set2)
print("Jaccard Distance: ", jaccard_dist)
```

Performing Distance Metrics on Uber fare Dataset

```python
data = pd.read_csv('uber.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
%pip install matplotlib seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plot (Two features)
plt.figure()
plt.scatter(df['fare_amount'],df['passenger_count'])
plt.xlabel('Fare Amount')
plt.ylabel('Passenger Count')
plt.title('Scatter Plot of Fare Amount vs Passenger Count')
plt.show()
```

```python
# Pair plot (all features)
sns.pairplot(df)
plt.show()
```

```python
# Distance Metrics Demonstration

# scaling using standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(df['fare_amount'].values.reshape(-1, 1))
print("Scaled data:\n", X_scaled)
Y_scaled = scaler.fit_transform(df['passenger_count'].values.reshape(-1, 1))
print("Scaled data:\n", Y_scaled)
```

```python
# Euclidean Distance
euclidean_dist = distance.euclidean(X_scaled[0], Y_scaled[0])
print("Euclidean Distance: ", euclidean_dist)

# Manhattan Distance
manhattan_dist = distance.cityblock(X_scaled[0], Y_scaled[0])   
print("Manhattan Distance: ", manhattan_dist)

# Cosine Similarity
cosine_sim = 1 - distance.cosine(X_scaled[0], Y_scaled[0])
print("Cosine Similarity: ", cosine_sim)

# Minkowski Distance (p=3 for demonstration)
minkowski_dist = distance.minkowski(X_scaled[0], Y_scaled[0], 3)
print("Minkowski Distance (p=3): ", minkowski_dist)

# Chessboard Distance
chessboard_dist = distance.chebyshev(X_scaled[0], Y_scaled[0])
print("Chessboard Distance: ", chessboard_dist)
```

Since this uber fare dataset does not contain any categorical data, hamming and jaccard distance metrics cannot be applied
